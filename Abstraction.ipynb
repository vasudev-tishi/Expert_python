# Abstraction-
 **The process by which data and functions are defined in such a way that only essential details can be seen and unnecessary implementation are hidden is called Data Abstraction.**

 How abstraction achieved in python:



*   By using abs module - ABC class
                           abstractmethod     
*   Inherit your class from ABC class

*   Create abstract methods in your abstract class

Syntax:
 from abc import ABC, abstractmethod
 class Employee(ABC)
        #abstract methods - method that has a declaration but does not have an implementation
       

  Abstract Classes (A class that cannot be instantiated)
  Abstract Methods (Methods that must be implemented by subclasses)




```python
from abc import ABC, abstractmethod #abc-absract base class

# Abstract class
class Animal(ABC):
    @abstractmethod #decorator symbol (signature)
    def make_sound(self):
        pass  # Abstract method with no implementation

# Subclass inheriting from the abstract class
class Dog(Animal):
    def make_sound(self):  #implementation
        print("Bark")

class Cat(Animal):
    def make_sound(self): #implementation
        print("Meow")

# Creating instances of subclasses
dog = Dog()
dog.make_sound()  # Output: Bark

cat = Cat()
cat.make_sound()  # Output: Meow
```

Explanation:
1. The Animal class is an abstract class because it inherits from ABC and has an abstract method make_sound().
2. The make_sound() method is defined but not implemented in the Animal class.
3. The Dog and Cat classes inherit from Animal and provide their own implementation of make_sound().
4. Trying to instantiate Animal directly will result in an error:

Key Points:
1. Abstract classes help in achieving data abstraction by hiding implementation

details.
2. Subclasses must implement the abstract methods to become concrete classes.
3. It enforces a structure, ensuring all derived classes implement specific methods.

```python
from abc import ABC, abstractmethod

class Car(ABC):
    @abstractmethod
    def mileage (self):
      pass
    def color(self):
      print("white")

class Suzuki(Car):
      def mileage(self):
        print("The mileage is 25kmph")

class Honda(Car):
      def mileage(self):
        print("The mileage is 24kmph")

s1 = Suzuki()
s1.mileage()
s1.color()
h1 = Honda()
h1.mileage()
h1.color()
```

```python
#Abstract Class for Vehicles

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def max_speed(self):
        pass

class Car(Vehicle):
    def max_speed(self):
        return 200

class Bike(Vehicle):
    def max_speed(self):
        return 100

car = Car()
print(car.max_speed())  # Output: 200

bike = Bike()
print(bike.max_speed())  # Output: 100
```

```python
from abc import ABC, abstractmethod #abc-absract base class

class BankApp(ABC):
  def database(self):
    print('connected to database')

  @abstractmethod
  def security(self):
    pass

class MobileApp(BankApp):
  def mobile_login(self):
    print('login into mobile')

  def security(self):
    print('mobile security')

mob = MobileApp()
mob.database()
mob.mobile_login()
mob.security()

```

```python
#Abstract Class for Shapes

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

circle = Circle(5)
print(circle.area())
```
