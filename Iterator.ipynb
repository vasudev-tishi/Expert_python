An **Iterator** in Python is an object that allows sequential access to elements of a collection (like lists, tuples, or dictionaries) without exposing the underlying structure. Iterators provide a way to access elements one at a time and remember their state between iterations.

Iteration - Process of taking each item of something one after another
example- Whenever you use a loop for visiting item of sequence

Every iterable is iterable. you can run loop on every iterator.

Create an iterator from a list and print its first two elements using next().

```python
# Creating a simple iterator using iter() and next()
numbers = [1, 2, 3, 4, 5]  # Iterable (list)
iterator = iter(numbers)  # Creating an iterator

print(next(iterator))   # Iteration
print(next(iterator))
print(next(iterator))   # Iteration


```

Convert a string into an iterator and print each character using next().

```python
text = "Hello"
text_iter = iter(text)  # Convert string into an iterator

print(next(text_iter))
print(next(text_iter))
print(next(text_iter))
```

By using for loop

```python
data = [1,2,3,4,5,6]

iterator = iter(data)

for item in iterator:
    print(item)
```

How to you use a for loop to iterate over an iterator.

```python
letters = iter(["a", "b", "c"])

for letter in letters:
    print(letter)  # What will this print?
```

How do you use iter() and next() to iterate through a range of numbers?

```python
range_iter = iter(range(5))

print(next(range_iter))
print(next(range_iter))
print(next(range_iter))
```

Manually iterate through a tuple using an iterator and a while loop.

```python
fruits = ("apple", "banana", "cherry")
fruit_iter = iter(fruits)

while True:
    try:
        print(next(fruit_iter))
    except StopIteration:
        break
```
